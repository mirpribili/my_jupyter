```python
# Магические методы __str__ и __repr__
# double underscore === dunder method's

class Lion(object):
	"""docstring for Lion"""
	def __init__(self, name):
		#super(Lion, self).__init__()
		self.name = name
		
	def __repr__(self): # как видит наш класс разработчик
		return f"The object Lion - {self.name}"

	def __str__(self): # как видит наш класс пользователь
		return f"Lion - {self.name}"



s = Lion('Simba')

print(s)

```

```python
Перегрузка операторов


Перегрузка операторов — один из способов реализации полиморфизма, когда мы можем задать свою реализацию какого-либо метода в своём классе.

Например, у нас есть два класса:

class A:
    def go(self):
        print('Go, A!')

class B(A):
    def go(self, name):
        print('Go, {}!'.format(name))

В данном примере класс B наследует класс A, но переопределяет метод go, поэтому он имеет мало общего с аналогичным методом класса A.

Однако в python имеются методы, которые, как правило, не вызываются напрямую, а вызываются встроенными функциями или операторами.

Например, метод __init__ перегружает конструктор класса. Конструктор - создание экземпляра класса.
>>>

>>> class A:
...     def __init__(self, name):
...         self.name = name
...
>>> a = A('Vasya')
>>> print(a.name)
Vasya

Собственно, далее пойдёт список таких "магических" методов.

__new__(cls[, ...]) — управляет созданием экземпляра. В качестве обязательного аргумента принимает класс (не путать с экземпляром). Должен возвращать экземпляр класса для его последующей его передачи методу __init__.

__init__(self[, ...]) - как уже было сказано выше, конструктор.

__del__(self) - вызывается при удалении объекта сборщиком мусора.

__repr__(self) - вызывается встроенной функцией repr; возвращает "сырые" данные, использующиеся для внутреннего представления в python.

__str__(self) - вызывается функциями str, print и format. Возвращает строковое представление объекта.

__bytes__(self) - вызывается функцией bytes при преобразовании к байтам.

__format__(self, format_spec) - используется функцией format (а также методом format у строк).

__lt__(self, other) - x < y вызывает x.__lt__(y).

__le__(self, other) - x ≤ y вызывает x.__le__(y).

__eq__(self, other) - x == y вызывает x.__eq__(y).

__ne__(self, other) - x != y вызывает x.__ne__(y)

__gt__(self, other) - x > y вызывает x.__gt__(y).

__ge__(self, other) - x ≥ y вызывает x.__ge__(y).

__hash__(self) - получение хэш-суммы объекта, например, для добавления в словарь.

__bool__(self) - вызывается при проверке истинности. Если этот метод не определён, вызывается метод __len__ (объекты, имеющие ненулевую длину, считаются истинными).

__getattr__(self, name) - вызывается, когда атрибут экземпляра класса не найден в обычных местах (например, у экземпляра нет метода с таким названием).

__setattr__(self, name, value) - назначение атрибута.

__delattr__(self, name) - удаление атрибута (del obj.name).

__call__(self[, args...]) - вызов экземпляра класса как функции.

__len__(self) - длина объекта.

__getitem__(self, key) - доступ по индексу (или ключу).

__setitem__(self, key, value) - назначение элемента по индексу.

__delitem__(self, key) - удаление элемента по индексу.

__iter__(self) - возвращает итератор для контейнера.

__reversed__(self) - итератор из элементов, следующих в обратном порядке.

__contains__(self, item) - проверка на принадлежность элемента контейнеру (item in self).
Перегрузка арифметических операторов

__add__(self, other) - сложение. x + y вызывает x.__add__(y).

__sub__(self, other) - вычитание (x - y).

__mul__(self, other) - умножение (x * y).

__truediv__(self, other) - деление (x / y).

__floordiv__(self, other) - целочисленное деление (x // y).

__mod__(self, other) - остаток от деления (x % y).

__divmod__(self, other) - частное и остаток (divmod(x, y)).

__pow__(self, other[, modulo]) - возведение в степень (x ** y, pow(x, y[, modulo])).

__lshift__(self, other) - битовый сдвиг влево (x << y).

__rshift__(self, other) - битовый сдвиг вправо (x >> y).

__and__(self, other) - битовое И (x & y).

__xor__(self, other) - битовое ИСКЛЮЧАЮЩЕЕ ИЛИ (x ^ y).

__or__(self, other) - битовое ИЛИ (x | y).

Пойдём дальше.

__radd__(self, other),

__rsub__(self, other),

__rmul__(self, other),

__rtruediv__(self, other),

__rfloordiv__(self, other),

__rmod__(self, other),

__rdivmod__(self, other),

__rpow__(self, other),

__rlshift__(self, other),

__rrshift__(self, other),

__rand__(self, other),

__rxor__(self, other),

__ror__(self, other) - делают то же самое, что и арифметические операторы, перечисленные выше, но для аргументов, находящихся справа, и только в случае, если для левого операнда не определён соответствующий метод.

Например, операция x + y будет сначала пытаться вызвать x.__add__(y), и только в том случае, если это не получилось, будет пытаться вызвать y.__radd__(x). Аналогично для остальных методов.

Идём дальше.

__iadd__(self, other) - +=.

__isub__(self, other) - -=.

__imul__(self, other) - *=.

__itruediv__(self, other) - /=.

__ifloordiv__(self, other) - //=.

__imod__(self, other) - %=.

__ipow__(self, other[, modulo]) - **=.

__ilshift__(self, other) - <<=.

__irshift__(self, other) - >>=.

__iand__(self, other) - &=.

__ixor__(self, other) - ^=.

__ior__(self, other) - |=.

__neg__(self) - унарный -.

__pos__(self) - унарный +.

__abs__(self) - модуль (abs()).

__invert__(self) - инверсия (~).

__complex__(self) - приведение к complex.

__int__(self) - приведение к int.

__float__(self) - приведение к float.

__round__(self[, n]) - округление.

__enter__(self), __exit__(self, exc_type, exc_value, traceback) - реализация менеджеров контекста.

Рассмотрим некоторые из этих методов на примере двухмерного вектора, для которого переопределим некоторые методы:

import math

class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return 'Vector2D({}, {})'.format(self.x, self.y)

    def __str__(self):
        return '({}, {})'.format(self.x, self.y)

    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)

    def __iadd__(self, other):
        self.x += other.x
        self.y += other.y
        return self

    def __sub__(self, other):
        return Vector2D(self.x - other.x, self.y - other.y)

    def __isub__(self, other):
        self.x -= other.x
        self.y -= other.y
        return self

    def __abs__(self):
        return math.hypot(self.x, self.y)

    def __bool__(self):
        return self.x != 0 or self.y != 0

    def __neg__(self):
        return Vector2D(-self.x, -self.y)

>>> x = Vector2D(3, 4)
>>> x
Vector2D(3, 4)
>>> print(x)
(3, 4)
>>> abs(x)
5.0
>>> y = Vector2D(5, 6)
>>> y
Vector2D(5, 6)
>>> x + y
Vector2D(8, 10)
>>> x - y
Vector2D(-2, -2)
>>> -x
Vector2D(-3, -4)
>>> x += y
>>> x
Vector2D(8, 10)
>>> bool(x)
True
>>> z = Vector2D(0, 0)
>>> bool(z)
False
>>> -z
Vector2D(0, 0)
```

## Пространства имен класса
[ООП 14 Пространство имен класса Class Body scope in Python](https://www.youtube.com/watch?v=vMS82GUSKik&list=PLQAt0m1f9OHvyjJNjZK_unnLwMOXPTja8&index=14)

```python
PYTHON_DEV = 0



# ----------------- GET
class DepartmentIT(object):
	"""docstring for DepartmentIT"""
	PYTHON_DEV = 10

	def info_out_of_class(self):
		print(PYTHON_DEV)

	def info_self(self):
		print(self.PYTHON_DEV)

	def info_ClassName(self):
		print(DepartmentIT.PYTHON_DEV)

	@property
	def info_property(self):
		print(self.PYTHON_DEV)


	@classmethod
	def info_classmethod(cls):
		print(cls.PYTHON_DEV)

	@staticmethod
	def info_staticmethod():
		print(DepartmentIT.PYTHON_DEV)

it = DepartmentIT()

it.info_out_of_class()
it.info_self()
it.info_ClassName()
it.info_property	
it.info_classmethod()
it.info_staticmethod()

print('-'*20)
# ----------------- SET


class DepartmentIT(object):
	"""docstring for DepartmentIT"""
	PYTHON_DEV = 10

	def info_self(self):
		print(self.PYTHON_DEV)

	def info_ClassName(self):
		print(DepartmentIT.PYTHON_DEV)

	def set_variable(self):
		self.PYTHON_DEV = self.PYTHON_DEV + 1



it = DepartmentIT()
it.info_self()
it.set_variable()
print(it.PYTHON_DEV) #  11 
print(DepartmentIT.PYTHON_DEV) # 10
# 10 != 11 !!!!!!!!!!!

print('-'*20)
# ----------------- SET2


class DepartmentIT(object):
	"""docstring for DepartmentIT"""
	PYTHON_DEV = 10

	def info_self(self):
		print(self.PYTHON_DEV)

	def info_ClassName(self):
		print(DepartmentIT.PYTHON_DEV)

	def set_variable(self):
		DepartmentIT.PYTHON_DEV = DepartmentIT.PYTHON_DEV + 1



it = DepartmentIT()
it.info_self()
it.set_variable()
print(it.__dict__)
print(it.PYTHON_DEV) #  11 
print(DepartmentIT.PYTHON_DEV) # 10
# 10 != 11 !!!!!!!!!!!
```

```python
class Vector:
    MIN_COORD = 0    
    MAX_COORD = 100
 
    def setCoords(self, x, y):
        if Vector.validate(x) and Vector.validate(y):
            self.x = x
            self.y = y
 
    @classmethod
    def validate(cls, arg):
        if arg >= cls.MIN_COORD and arg <= cls.MAX_COORD:
            return True
        return False
 
    @staticmethod
    def norm2(x, y):
        return x*x + y*y

#И для удобства открою консоль Python, где импортирую этот класс из модуля ex1:

#from ex1 import Vector

#Далее, смотрите, непосредственно через класс Vector можно вызвать метод класса:

Vector.validate(5)

#или, статический метод:

Vector.norm2(1, 2)

#Но нельзя вызвать обычный метод:

#Vector.setCoords(1, 2)


v = Vector()

#Тогда можно вызвать setCoords, либо через экземпляр класса:

v.setCoords(1, 2)

#либо, через сам класс, но уже с явной передачей первого параметра – ссылки на ранее созданный объект:

Vector.setCoords(v, 10, 20)

## ЗАЧЕМ?

#Если нам нужен метод, который бы «работал» с атрибутами экземпляров классов, 
#  то это однозначно обычный метод класса с первым параметром self, который указывает на текущий объект.

#Если нужен метод, который можно вызывать непосредственно из класса (или экземпляра)
#  и, который бы имел доступ к свойствам и методам этого класса, то его следует объявить 
#    как метод класса через декоратор @classmethod.

#Если нужен метод, который можно вызывать непосредственно из класса, 
#  но доступ к его атрибутам не предполагается, то достаточно его объявить 
#    как статический через декоратор @staticmethod. 
```

```python
from functools import wraps
def table(func):

    @wraps(func)
    def inner(*args, **kwargs):
        print("<table>")
        func(*args, **kwargs)
        print("</table>")
    return inner


def say():
    print("go home and stay home")

say = table(say)
print("say.__name__:\t\t",say.__name__)
print("say:\t\t", say)
print("say():\t\t",say())



  
def sqr(x):
    """

    Функция возводит в квадрат х
    :param x:
    :return:
    """
    print(x**2)
print("sqr.__name__:\t\t",sqr.__name__)
print("sqr.__doc__:\t\t",sqr.__doc__)
help(sqr)

sqr = table(sqr)

print("sqr.__name__:\t\t",sqr.__name__)
print("sqr.__doc__:\t\t",sqr.__doc__)
help(sqr)
```

```python
import turtle  # импортируем модуль Черепашка
import random  # импортируем модуль для работы со случайными числами

# Экран 
window = turtle.Screen()  # получение объекта экрана
window.bgcolor("black")  # изменение цвета окна
window.tracer(2)  # скорость обновления экрана
window.bgpic("game-bg.gif")

# Рамка поля
frame = turtle.Turtle()  # создание черепашки
frame.penup()  # поднимаем перо
frame.goto(300, 300)  # переходим в правый верхний угол
frame.width(3)  # увеличиваем толщину пера
frame.pendown()  # опускаем перо
frame.color("white")  # меняем цвет пера
for i in range(4):  # 4 раза делаем в цикле:
    frame.right(90)  # поворот направо на 90 градусов
    frame.forward(600)  # идем вперед на 600 пикселей
frame.hideturtle()  # прячем черепашку

# Игрок
player = turtle.Turtle()  # создаем черепашку игрока
player.color("blue")  # меняем цвет
player.shape("triangle")  # меняем форму на треугольник
player.penup()  # поднимаем перо
angle = random.randint(0, 360)  # выбираем случайный угол
player.right(angle)  # поворачиваемся на случайный угол

# Цели / еда
goals = []  # пустой список для целей
for i in range(10):  # 10 раз в цикле
    goal = turtle.Turtle()  # создаем одну цель
    goal.color("red")  # меняем цвет
    goal.shape("circle")  # меняем форму на круг
    goal.penup()  # поднимаем перо

    # выбираем случайную координату и переходим в нее
    x = random.randint(-290, 290)
    y = random.randint(-290, 290)
    goal.goto(x, y)

    # выбираем случайный угол и поворачиваемся на него
    angle = random.randint(0, 360)
    goal.right(angle)

    goals.append(goal)  # добавляем новую цель в список целей


def turn_left():
    player.left(20)


def turn_right():
    player.right(20)


def collide(t1, t2):  # касаются ли две черепашки
    dist = t1.distance(t2)  # расстояние между черепашками
    if dist < 20:
        return True  # касаются
    else:
        return False  # не касаются


# Обрабатывать клавиши (действия пользователя)
window.listen()
window.onkey(turn_left, "Left")  # при нажатии на клавишу `Left`  выполнить функцию
window.onkey(turn_right, "Right")  # при нажатии на клавишу `Right` выполнить функцию

info = turtle.Turtle()
info.color("white")
info.hideturtle()      # прячем
info.penup()           # поднимаем перо
info.goto(0, 310)      # в верхнюю часть экрана

# Основной цикл: движения, потом правила
speed = 1
score = 0
while True:  # бесконечный цикл
    player.forward(speed)  # движение игрока

    if player.xcor() > 290 or player.xcor() < -290:  # отталкивание от стенок по оси Х
        player.right(180)
    if player.ycor() > 290 or player.ycor() < -290:  # отталкивание от стенок по оси Y
        player.right(180)

    for g in goals:  # для всех целей
        g.forward(speed)  # движение целей

        if g.xcor() > 290 or g.xcor() < -290:  # отталикание от стенок X
            g.right(180)
        if g.ycor() > 290 or g.ycor() < -290:  # отталкивание от стенок Y
            g.right(180)

        if collide(player, g):
            # обновление счета
            score = score + 1
            info.clear()       # очищаем все написанное  # выбираем случайную координату и переходим в нее
            info.write(score, align="center", font=("Arial", 14)  )
            # выбираем случайную координату и переходим в нее
            x = random.randint(-290, 290)
            y = random.randint(-290, 290)
            g.goto(x, y)

            # выбираем случайный угол и поворачиваемся на него
            angle = random.randint(0, 360)
            g.right(angle)

turtle.exitonclick()  # закрытие только по команде пользователя
```

# Черепашьи гонки

```python

import turtle
import random

field = turtle.Turtle()
field.penup()
field.goto(-200, 200)
field.pendown()
field.speed(10)

line = 300

for i in range(8):
    field.right(90)
    field.forward(line)
    field.left(90)
    field.penup()
    field.forward(line / 10)
    field.pendown()
    field.left(90)
    field.forward(line)
    field.right(90)
    field.penup()
    field.forward(line / 10)
    field.pendown()
field.hideturtle()

t1 = turtle.Turtle()
t1.shape('turtle')
t1.color('red')
t1.penup()
t1.goto(-200, 50)

t2 = turtle.Turtle()
t2.shape('turtle')
t2.color('blue')
t2.penup()
t2.goto(-200, -50)

finish = -200 + 15 * (line / 10)

while t1.xcor() < finish and t2.xcor() < finish:
    step1 = random.randint(1, 5)
    step2 = random.randint(1, 5)
    t1.forward(step1)
    t2.forward(step2)

def dance(t):
    t.pendown()
    t.goto(0, 100)
    for i in range(4):
        t.forward(25)
        t.left(90)


if t1.xcor() >= finish:
    print("t1 win!")
    dance(t1)
elif t2.xcor() >= finish:
    print("t2 win!")
    dance(t2)
else: # если у вас три черепашки
    print("t3 win!")
    dance(t3)


turtle.exitonclick()
```
